```python
import os
import pandas as pd
import sqlite3
import numpy as np
```

```python
path1 = "/content/drive/MyDrive/DS340W_OilFutures/Data/WTI 2004-2014.csv"

path2 = "/content/drive/MyDrive/DS340W_OilFutures/Data/WTI 2014-2026.csv"

path3 = "/content/drive/MyDrive/DS340W_OilFutures/Data/WTI 2000-2003.csv"

DF1 = pd.read_csv(path1)

DF2 = pd.read_csv(path2)

DF3 = pd.read_csv(path3)

DF1
```

```python
DF1 = pd.read_csv(path1)
DF2 = pd.read_csv(path2)
DF3 = pd.read_csv(path3)

DF1['Date'] = pd.to_datetime(DF1['Date'], format='%m/%d/%Y')
DF2['Date'] = pd.to_datetime(DF2['Date'], format='%m/%d/%Y')
DF3['Date'] = pd.to_datetime(DF3['Date'], format='%m/%d/%Y')

df_all = (
    pd.concat([DF1, DF2, DF3], ignore_index=True)
      .drop_duplicates(subset=['Date'])
      .sort_values('Date', ascending=True)
      .reset_index(drop=True)
)

df_all.head(), df_all.tail()
```

```python
df_all = (
    df_all
        .sort_values('Date')     # safety check
        .set_index('Date')       # Date becomes the index
)
```

```python
save_path = "/content/drive/MyDrive/DS340W_OilFutures/Data/WTI 2000-2026.csv"

df_all.to_csv(save_path)
```

```python

df_lag = df_all.sort_index()

# Create lag features
for lag in range(1, 11):
    df_lag[f'Price_lag{lag}'] = df_lag['Price'].shift(lag)

df_lag.head(15)
```

```python
save_path = "/content/drive/MyDrive/DS340W_OilFutures/Data/WTI_10DayLag.csv"
df_lag.to_csv(save_path)
```

```python
open_path = "/content/drive/MyDrive/DS340W_OilFutures/Data/WTI_10DayLag.csv"
df_lag = pd.read_csv(open_path, index_col=0, parse_dates=True)
df_lag
```

```python
#remove rows with missing lag variables
df_lag = df_lag.dropna()
df_lag
```

70/20/10 split for training/validation/test sets

```python
n = len(df_lag)

train_end = int(n * 0.70)
val_end   = int(n * 0.90)

train_df = df_lag.iloc[:train_end]
val_df   = df_lag.iloc[train_end:val_end]
test_df  = df_lag.iloc[val_end:]
```

```python
WTI_Lag_Train = "/content/drive/MyDrive/DS340W_OilFutures/Data/WTI_Lag_Train.csv"
WTI_Lag_Val = "/content/drive/MyDrive/DS340W_OilFutures/Data/WTI_Lag_Val.csv"
WTI_Lag_Test = "/content/drive/MyDrive/DS340W_OilFutures/Data/WTI_Lag_Test.csv"

train_df.to_csv(WTI_Lag_Train)
val_df.to_csv(WTI_Lag_Val)
test_df.to_csv(WTI_Lag_Test)
```
